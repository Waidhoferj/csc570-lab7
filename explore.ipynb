```python
img_out = tf.ones((10,10), dtype="float32")
right_pad_shape = list(img_out.shape)
right_pad_shape[1] = 1
angle = tf.fill((3,),2.0)
target_pose = tf.fill((3,),3.0)
right_pad = tf.zeros((right_pad_shape),dtype="float32")
img = tf.concat([img_out, right_pad], axis=1)
pose_row = tf.concat([angle, target_pose, tf.zeros((img_out.shape[1] - 5,))], axis=0)
pose_row = tf.reshape(pose_row,(1,-1))
img = tf.concat([img, pose_row], axis=0)
img
```
